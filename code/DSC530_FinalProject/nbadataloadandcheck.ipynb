```python
import pandas as pd
import glob

#Path is location of NBA Game logs on my computer after downloading from repository
#Repository location : https://github.com/ryanvmenezes/lawlers-law/tree/master/data/logs

path = r'C:\Users\Danny\Documents\data\proj' # use your path
all_files = glob.glob(path+ "\*.csv")

#Header refers to the position of Column Headers
#Axis determines whether it will be merged on rows or columns
frame = pd.concat(((pd.read_csv(f, index_col=None, header=0)) for f in all_files)
                  , axis=0, ignore_index=True)
```

```python
frame.to_csv( "nbagames1985to2019.csv", index=False, encoding='utf-8-sig')
```
